## E. Data Preparation

### E.1 Data Transformation add leakage


```python
def make_feature_versions(df, features_list, leak_prefixes):
    # core feature
    cols = [c for c in df.columns if any(c.startswith(p) for p in features_list)]
    df_selected = df[cols].copy()
    # print(features_list)
    # safe version
    df_safe = df_selected.drop(columns=[c for c in df.columns if any(c.startswith(p) for p in leak_prefixes)], errors="ignore")
    # with leak ver
    df_leak = df_selected.copy()
    return df_safe, df_leak
```

```python
X_train_safe, X_train_leak = make_feature_versions(X_train, features_list + leak_cols, leak_cols)
X_val_safe,   X_val_leak   = make_feature_versions(X_val,   features_list + leak_cols, leak_cols)
X_test_safe,  X_test_leak  = make_feature_versions(X_test,  features_list + leak_cols, leak_cols)

print("Safe version shape:", X_train_safe.shape)
print("Leakage version shape:", X_train_leak.shape)
```

```python
# Provide some explanations on why you believe it is important to perform this data transformation and its impacts

data_transformation_1_explanations = """
We use the same train/val/test split (70/15/15, stratified). Categorical features are one-hot encoded with unknowns ignored. Numeric features use light clipping for extreme tails. Transformers are fit on train only to avoid leakage. No scaling is required for RF.
"""
```

### G.2 Set Hyperparameters

```python
# Random Search Hyperparameters

param_dist = {
    "n_estimators": randint(300, 600),
    "max_depth": [None, 8, 12, 16],
    "min_samples_leaf": randint(1, 5),
    "max_features": ["sqrt", "log2"],
    "class_weight": [None, "balanced"]
}

rand_search = RandomizedSearchCV(
    RandomForestClassifier(random_state=42, n_jobs=-1),
    param_distributions=param_dist,
    n_iter=30,   #  30 features chosen
    scoring="f1_macro",
    cv=5,
    n_jobs=-1,
    verbose=1,
    random_state=42
)

rand_search.fit(X_train_safe, y_train_enc)
print(rand_search.best_params_, rand_search.best_score_)
```

```python
# Grid Search Hyperparameters
param_grid = {
    "n_estimators": [300, 400, 500],
    "max_depth": [16, 18, 20],
    # fix the following based on the random search best
    "min_samples_leaf": [2, 3],
    "max_features": ["sqrt", "log2"],
    "class_weight": ["balanced"],
    # optional: keep bootstrap=True
}

# 3) CV setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

rf = RandomForestClassifier(
    random_state=42,
    n_jobs=-1,
    bootstrap=True
)

grid = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    scoring={"macro_f1": "f1_macro", "bacc": "balanced_accuracy"},
    refit="macro_f1",     # use macro-F1 to refit
    cv=cv,
    n_jobs=-1,
    verbose=1
)

grid.fit(X_train_safe, y_train_enc)

print("Best params (fine-tune):", grid.best_params_)
print("CV best macro-F1:", f"{grid.best_score_:.4f}")

best_rf = grid.best_estimator_
```

```python
# Step 1: narrow grid on depth & leaf
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

param_grid_cap = {
    "max_depth": [15, 16, 17, 18],
    "min_samples_leaf": [2, 3],
}

rf_base = RandomForestClassifier(
    n_estimators=500,              # temporary fix
    max_features="sqrt",
    class_weight="balanced",
    bootstrap=True,
    n_jobs=-1,
    random_state=42
)

g_cap = GridSearchCV(
    rf_base, param_grid_cap,
    scoring="f1_macro", cv=5, n_jobs=-1, verbose=1
)
g_cap.fit(X_train_safe, y_train_enc)
print("Best capacity params:", g_cap.best_params_, "CV macro-F1:", f"{g_cap.best_score_:.4f}")

best_depth = g_cap.best_params_["max_depth"]
best_leaf  = g_cap.best_params_["min_samples_leaf"]
```

```python
# Step 2: 1D sweep on n_estimators
from sklearn.model_selection import cross_val_score
import numpy as np

n_list = [450, 500, 550]
records = []

for n in n_list:
    rf_tmp = RandomForestClassifier(
        n_estimators=n,
        max_depth=best_depth,
        min_samples_leaf=best_leaf,
        max_features="sqrt",
        class_weight="balanced",
        bootstrap=True,
        n_jobs=-1,
        random_state=42
    )
    cv_scores = cross_val_score(rf_tmp, X_train_safe, y_train_enc, cv=5, scoring="f1_macro", n_jobs=-1)
    records.append((n, np.mean(cv_scores), np.std(cv_scores)))

for n, m, s in records:
    print(f"n_estimators={n} | CV macro-F1={m:.4f} ± {s:.4f}")
```

```python
# Stability check across random_state
final_params = {
    "n_estimators": 500,
    "max_depth": 17,
    "min_samples_leaf": 3,
    "max_features": "sqrt",
    "class_weight": "balanced",
    "bootstrap": True,
    "n_jobs": -1,
    "random_state": 42
}

seeds = [33, 2023, 7]

for s in seeds:
    rf_final = RandomForestClassifier(
        n_estimators=500,
        max_depth=best_depth,
        min_samples_leaf=best_leaf,
        max_features="sqrt",
        class_weight="balanced",
        bootstrap=True,
        n_jobs=-1,
        random_state=s
    )
    rf_final.fit(X_train_safe, y_train_enc)
    val_pred = rf_final.predict(X_val_safe)
    v_macro  = f1_score(y_val_enc, val_pred, average="macro")
    v_bacc   = balanced_accuracy_score(y_val_enc, val_pred)
    print(f"seed={s} | Val Macro-F1={v_macro:.4f} | Val bAcc={v_bacc:.4f}")
```

```python
# Grid Search Hyperparameters
param_grid_leak = {
    "n_estimators": [300, 400, 500],
    "max_depth": [16, 18, 20],
    # fix the following based on the random search best
    "min_samples_leaf": [1, 2, 3],
    "max_features": ["sqrt", "log2"],
    "class_weight": ["balanced"],
    # keep bootstrap=True
}

# 3) CV setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

rf_leak = RandomForestClassifier(
    random_state=42,
    n_jobs=-1,
    bootstrap=True
)

grid_leak = GridSearchCV(
    estimator=rf_leak,
    param_grid=param_grid_leak,
    scoring={"macro_f1": "f1_macro", "bacc": "balanced_accuracy"},
    refit="macro_f1",     # use macro-F1 to refit
    cv=cv,
    n_jobs=-1,
    verbose=1
)

grid_leak.fit(X_train_leak, y_train_enc)

print("Best params (fine-tune):", grid_leak.best_params_)
print("CV best macro-F1:", f"{grid_leak.best_score_:.4f}")

best_rf_leak = grid_leak.best_estimator_
```
